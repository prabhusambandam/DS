```python
import pandas as pd

# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
df = pd.read_csv("sample.csv",names=["X1","X2","Y"] )
```

```python
df
```

```python
model = LinearRegression()
```

```python
model.fit(df[["X1"]], df["Y"])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
df.plot(kind = "scatter", x = "X1", y = "Y")
```

```python
pred_df = model.predict(df[["X1"]])
```

```python

plt.scatter(df.X1,df.Y, color = "red")
plt.plot(df.X1,pred_df, 'b' )
```

```python
model = LinearRegression()

model.fit(df[["X1"]], df["Y"])

print("Slope = ",model.coef_)

print("Intercept = ",np.ceil(model.intercept_))

df.plot(kind = "scatter", x = "X1", y = "Y")

pred_df = model.predict(df[["X1"]])


plt.scatter(df.X1,df.Y, color = "red")
plt.plot(df.X1,pred_df, 'b' )
```

```python
model = Lasso(alpha=0)

model.fit(df[["X1"]], df["Y"])

print("Slope = ",model.coef_)

print("Intercept = ",np.ceil(model.intercept_))

df.plot(kind = "scatter", x = "X1", y = "Y")

pred_df = model.predict(df[["X1"]])


plt.scatter(df.X1,df.Y, color = "red")
plt.plot(df.X1,pred_df, 'b' )
```

```python
model = Lasso(alpha=0.024)

model.fit(df[["X1"]], df["Y"])

print("Slope = ",model.coef_)

print("Intercept = ",np.ceil(model.intercept_))

df.plot(kind = "scatter", x = "X1", y = "Y")

pred_df = model.predict(df[["X1"]])


plt.scatter(df.X1,df.Y, color = "red")
plt.plot(df.X1,pred_df, 'b' )
```

```python
np.arange(0,0.05,0.001)
```

```python
for i in np.arange(0,0.05,0.001):
    print(i)
    model = Lasso(alpha=i)

    model.fit(df[["X1"]], df["Y"])

    print("Slope = ",model.coef_)

    print("Intercept = ",np.ceil(model.intercept_))

    df.plot(kind = "scatter", x = "X1", y = "Y")

    pred_df = model.predict(df[["X1"]])


    plt.scatter(df.X1,df.Y, color = "red")
    plt.plot(df.X1,pred_df, 'b' )
```

```python
model = Ridge(alpha=100005)

model.fit(df[["X1"]], df["Y"])

print("Slope = ",model.coef_)

print("Intercept = ",np.ceil(model.intercept_))

df.plot(kind = "scatter", x = "X1", y = "Y")

pred_df = model.predict(df[["X1"]])


plt.scatter(df.X1,df.Y, color = "red")
plt.plot(df.X1,pred_df, 'b' )
```
